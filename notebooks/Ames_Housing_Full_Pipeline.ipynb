```python
col_to_none=['Garage Finish', 'Garage Qual', 'Garage Cond', 
                    'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 
                    'BsmtFin Type 2','Garage Type']
for col in col_to_none:
    df[col]=df[col].fillna('None')
df['Mas Vnr Type'] = df['Mas Vnr Type'].fillna('None') 
df['Fireplace Qu']=df['Fireplace Qu'].fillna('None')   

num_cols_to_zero = ['Garage Area', 'Garage Cars', 'Mas Vnr Area', 'BsmtFin SF 1', 
                    'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']

for col in num_cols_to_zero:
    df[col] = df[col].fillna(0)

df['Lot Frontage'] = df['Lot Frontage'].fillna(df['Lot Frontage'].median())
df['Electrical']=df['Electrical'].fillna(df['Electrical'].mode()[0])   
```

```python
# Identify the nominal columns
nominal_cols = ['MS SubClass', 'MS Zoning', 'Street', 'Land Contour', 'Lot Config', 
                'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 
                'Roof Style', 'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 
                'Foundation', 'Heating', 'Central Air', 'Garage Type', 'Sale Type', 'Sale Condition']

# Convert MS SubClass to string first so it's treated as a category
df['MS SubClass'] = df['MS SubClass'].astype(str)

# Create Dummy Variables
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```

```python
# Lot Shape: Reg (Regular) is usually 'better' than Irregular for many buyers
shape_map = {'Reg': 4, 'IR1': 3, 'IR2': 2, 'IR3': 1}

# Bsmt Exposure: Gd (Good) > Av (Average) > Mn (Minimum) > No (No Exposure) > None
exposure_map = {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1, 'None': 0}

# BsmtFin Type: GLQ (Good Living Quarters) is the top rank
fin_map = {'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'None': 0}

# Functional: Typ (Typical) is the best/standard
func_map = {'Typ': 7, 'Min1': 6, 'Min2': 5, 'Mod': 4, 'Maj1': 3, 'Maj2': 2, 'Sev': 1, 'Sal': 0}

# Garage Finish: Fin (Finished) > RFn (Rough Finished) > Unf (Unfinished)
finish_map = {'Fin': 3, 'RFn': 2, 'Unf': 1, 'None': 0}

# Paved Drive: Y (Paved) > P (Partial) > N (Dirt/Gravel)
pave_map = {'Y': 3, 'P': 2, 'N': 1}

# Apply mappings
df['Lot Shape'] = df['Lot Shape'].map(shape_map)
df['Bsmt Exposure'] = df['Bsmt Exposure'].map(exposure_map)
df['BsmtFin Type 1'] = df['BsmtFin Type 1'].map(fin_map)
df['BsmtFin Type 2'] = df['BsmtFin Type 2'].map(fin_map)
df['Functional'] = df['Functional'].map(func_map)
df['Garage Finish'] = df['Garage Finish'].map(finish_map)
df['Paved Drive'] = df['Paved Drive'].map(pave_map)

# Create dummies for the last two columns
df = pd.get_dummies(df, columns=['Utilities', 'Electrical'], drop_first=True)
```

```python
# 1. Calculate correlations with SalePrice
correlations = df.corr()['SalePrice'].sort_values(ascending=False)

# 2. Look at the top 20 features that INCREASE price
print("--- Top Positive Predictors ---")
print(correlations.head(20))

# 3. Look at the top 10 features that DECREASE price
print("\n--- Top Negative Predictors ---")
print(correlations.tail(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score 
features=['Overall Qual', 'Gr Liv Area', 'Exter Qual', 'Kitchen Qual', 
'Total Bsmt SF', 'Garage Cars', 'Year Built', 'Foundation_PConc']
X=df[features]
Y=df['SalePrice']
#train test split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
#train the model 
model=LinearRegression()
model.fit(X_train,Y_train)
prediction=model.predict(X_test)
mae=mean_absolute_error(Y_test,prediction)
score=r2_score(Y_test,prediction)
print(mae)
print(score)


```

LassoCV model


```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_absolute_error,r2_score
X=df.drop('SalePrice',axis=1)
Y=df['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
# 3. Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
lasso=LassoCV(cv=5)
lasso.fit(X_train_scaled,Y_train)
prediction=lasso.predict(X_test_scaled)
mae=mean_absolute_error(Y_test,prediction)
score=r2_score(Y_test,prediction)
print(mae)
print(score)
```
